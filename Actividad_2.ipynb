```python
import pandas as pd
ruta = "/content/drive/MyDrive/Colab Notebooks/archivos para practica/"

# 1.
ventas = pd.read_csv(ruta + "ventas.csv", encoding="ISO-8859-1")
print("Primeras filas del archivo ventas:")
print(ventas.head())

# 2
clientes = pd.read_json(ruta + "clientes.json")
print("\nCantidad total de clientes:")
print(len(clientes))

# 3
inventario = pd.read_excel(ruta + "inventario.xlsx")
print("Columnas del inventario:", ", ".join(inventario.columns))

promedio_stock = inventario["stock_disponible"].mean()
print("\nPromedio de stock disponible:")
print(promedio_stock)
```
