```python
valor=np.random.randint(1,101,(15))

valormax=np.max(valor)
valormin=np.min(valor)
valorpro=np.average(valor)
print('Los valores son:',valor)
print('El valor maximo es:',valormax)
print('El valor minimo es:',valormin)
print('El valor promedio es:',valorpro)
```

```python
matriz=np.arange(1,26).reshape(5,5)
print(matriz)

diagonal=np.diag(matriz)
print('La diagonal esta compuesta por:',diagonal)
```

```python
matriz=np.random.randint(1,21,(5,5))
print(matriz)

sumafila=np.sum(matriz , axis=1)
print('La suma de las filas:',sumafila)
sumacol=np.sum(matriz , axis=0)
print('La suma de las columnas:',sumacol)
```
